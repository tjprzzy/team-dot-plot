```python
import pickle
import numpy as np
```

```python
infile = open("llbean_data_1.pkl",'rb')
data = pickle.load(infile)
```

# Cleaning the Author Field

```python
data["verified_purch"] = ""
data["feedback_count"] = ""
data["pos_feedback_count"] = ""
data["user"] = ""
data["user_location"] = ""

for i in range(len(data[3])):
    
    # Split the field up into a list using seperator = "\n"
    tmp = data[3][i].split("\n")
    
    # Remove useless text
    tmp.remove("Was this helpful?YesNo")
    tmp = [x for x in tmp if "Contributor" not in x]
    tmp = [x for x in tmp if "Featured" not in x]
    
    # Verified Purchase
    if 'Verified Purchase' in tmp:
        data["verified_purch"][i] = "Y"
    else:
        data["verified_purch"][i] = "N"    
    tmp = [x for x in tmp if "Verified Purchase" not in x]
    
    # Review Feedback Counts
    data["pos_feedback_count"][i] = [x[:x.find(' of ')] for x in tmp if "found this review" in x]
    if len(data["pos_feedback_count"][i])==1:
        data["pos_feedback_count"][i]=data["pos_feedback_count"][i][0]
    else:
        data["pos_feedback_count"][i]=0
    
    data["feedback_count"][i] = [x for x in tmp if "found this review" in x]
    if len(data["feedback_count"][i])==1:
        data["feedback_count"][i] = data["feedback_count"][i][0].replace(" found this review helpful.","").replace(data["pos_feedback_count"][i]+" of","")
    else:
        data["feedback_count"][i]=0
    
    tmp = [x for x in tmp if "found this review helpful" not in x]
    
    if len(tmp) == 2:
        data["user"][i] = tmp[0]
        data["user_location"][i] = tmp[1]
    else:
        data["user_location"][i] = tmp[0]

del data[3] 
```

# Cleaning the Size Field

```python
data["product_size"] = ""
data["product_fit"] = ""
data["user_fit"] = ""
data["user_height"] = ""
data["user_weight"] = ""

for i in range(len(data[4])):
    
    # Split the field up into a list using seperator = "\n"
    tmp = data[4][i].split("\n")
    
    # Defining product size
    data["product_size"][i] = [x for x in tmp if "Size:" in x]
    if len(data["product_size"][i])==1:
        data["product_size"][i] = data["product_size"][i][0].replace("Size: ","")
    else:
        data["product_size"][i]=""
    tmp = [x for x in tmp if "Size:" not in x]

    # Defining user fit
    data["user_fit"][i] = [x for x in tmp if "Fit" in x]
    if len(data["user_fit"][i])==1:
        data["user_fit"][i] = data["user_fit"][i][0].replace("Fit: ","")
    else:
        data["user_fit"][i]=""
    tmp = [x for x in tmp if "Fit:" not in x]

    # Defining the product fit line
    data["product_fit"][i] = [x for x in tmp if "Size Range: " in x]
    if len(data["product_fit"][i])==1:
        data["product_fit"][i] = data["product_fit"][i][0].replace("Size Range: ","")
    else:
        data["product_fit"][i]=""
    tmp = [x for x in tmp if "Size Range: " not in x]
    
    # Defining user weight
    data["user_weight"][i] = [x[:x.find(' lbs')] for x in tmp if "lbs" in x]
    if len(data["user_weight"][i])==1:
        data["user_weight"][i] = data["user_weight"][i][0][-4:].strip()
    else:
        data["user_weight"][i]=""   
    
    # Defining user weight
    data["user_height"][i] = [x[:1+x.find('\"')] for x in tmp if '\"' in x]
    if len(data["user_height"][i])==1:
        data["user_height"][i] = data["user_height"][i][0].replace("Customer is ","").strip()
    else:
        data["user_height"][i]=""

del data[4] 
```

# Clean the Rating Field

```python
data = data.rename(columns={data.columns[0]: 'description',data.columns[1]: 'review_title',data.columns[2]: 'review_date',data.columns[3]: 'review_text',data.columns[4]: 'rating'})
for i in range(len(data['rating'])):
    data['rating'][i] = data['rating'][i][:data['rating'][i].find(' out of')].replace("Rating: ","")

data 
```
